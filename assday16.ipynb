```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.relplot(x='parental level of education',y='math score',data=df, hue='gender',style='race/ethnicity',kind='line')
plt.xticks(rotation=20)
plt.title('RELPLOT')
plt.xlabel('P a r e n t a l   L e v e l   O f   E d u c a t i o n')
plt.ylabel('M a t h s   S c o r e')
plt.show()
```

```python
sns.catplot(x='gender',y='reading score',data=df,kind='violin',hue='race/ethnicity')
plt.title('CATPLOT')
plt.xlabel('G e n d e r')
plt.ylabel('R e a d i n g   S c o r e')
plt.show()
```

```python
sns.catplot(x='gender',y='reading score',data=df,kind='violin',hue='race/ethnicity',col='lunch')
plt.show()
```

```python
sns.scatterplot(x='parental level of education',y='writing score',data=df,hue='gender')
plt.xticks(rotation=30)
plt.title('SCATTER PLOT')
plt.xlabel('P a r e n t a l   L e v e l   O f   E d u c a t i o n')
plt.ylabel('W r i t i n g   S c o r e')
plt.show()
```

```python
sns.barplot(x='parental level of education',y='writing score',data=df,hue='gender')
plt.xticks(rotation=30)
plt.title('BAR PLOT')
plt.xlabel('P a r e n t a l   L e v e l   O f   E d u c a t i o n')
plt.ylabel('W r i t i n g   S c o r e')
plt.show()
```

```python
key=list(df['gender'].value_counts().keys())
key
```
